### Create Own Iterator

```python
class Nums:  # intialize , current value store, incresing value upto "stopIteration"
    MAX = 4
    def __init__(self):
        self.current = 0
    def __iter__(self):
        return self
    def __next__(self):
        next_value = self.current
        if next_value >= self.MAX:
            raise StopIteration
        self.current+=1
        return next_value

nums = Nums()
print(nums)
```

```python
for num in nums:
    print(num)
```

```python
# Infinate Iterator
class Example:
    def __iter__(self):
        self.a = 1
        return self
    def __next__(self):
        x = self.a
        self.a+=5
        return x

# Create Iterator/Class Object
temp = Example()
print(temp)
```

```python
for t in temp:
    print(t,end= " ")
```

```python
t1 = Example()
print(t1)
my_temp = iter(t1)
print(my_temp)
```

```python
print(next(my_temp))
```

```python
print(next(my_temp))
```

```python
print(next(my_temp))
```

## Generators
- Generator is a function that returns an Object which can iterate over as one value at a time.

#### Return Statement will terminate the entire function
#### Yield Pauses the function & saving all its sates

```python
# print numbers from 0 to 3 using Generator
def gen_nums():
    n = 0
    while n < 4:
        yield n
        n += 1

gen_temp = gen_nums()
print(gen_temp)
        
```

```python
print(list(gen_temp))
```

```python
for m in gen_temp:
    print(m,end= ' ')
```

```python
print(next(gen_temp))
```

```python
print(next(gen_temp))
```

```python
print(next(gen_temp))
```

```python
print(next(gen_temp))
```

### Benfits of Generators
- Simplified Code
- Better Performance

```python
# Example -2 
def mygen1(n):
    for  i in range(n):
        yield i
    
def mygen2(n,m):
    for j in range(n,m):
        yield j
        
def mygen3(n,m):
    yield from mygen1(n)
    yield from mygen2(n,m)
    yield from mygen2(n,m)

print(list(mygen1(5)))
print(list(mygen2(5,10)))
print(list(mygen3(0,10)))
```

## Lambda
- Lambda is a small anonymous Function
- Lambda function can take any number of arguments, but can only have one Expression

```python
# Syntax of Lambda:-
#  lambda arguments : expression
```

```python
x= lambda a: a+10
print(x(5))
```

```python
# Multiplication of a,b
mul = lambda a,b : a*b
mul(8,9)
```

```python
add =lambda x,y,z : x+y+z
add(1,2,3)
```
